```python
#Stochastic Gradient Descent
import numpy as np
import matplotlib.pyplot as plt

# Generating synthetic data
x = np.random.rand(100, 1)  # 100 random numbers between 0 and 1
y = 2 * x + 1 + np.random.randn(100, 1)  # y = 2x + 1 + noise

# Initializing parameters
w = np.random.rand(1)
b = np.random.rand(1)
learning_rate = 0.01

# Defining the SGD update rule
def sgd_update(xi, yi, w, b, learning_rate):
    yhat = w * xi + b
    error = yi - yhat
    dw = -2 * xi * error
    db = -2 * error
    w = w - learning_rate * dw
    b = b - learning_rate * db
    cost = error ** 2
    return w, b, cost

# Training using SGD
epochs = 50
arr = []

for epoch in range(epochs):
    for i in range(len(x)):
        w, b, cost = sgd_update(x[i], y[i], w, b, learning_rate)
    
    # Calculate the total loss for the current epoch
    yhat = w * x + b
    loss = np.mean((y - yhat) ** 2)
    print(f'Epoch {epoch} loss is {loss}, parameters are w = {w}, b = {b}')
    arr.append([epoch, loss])

# Plotting the loss over epochs
plt.plot([i[0] for i in arr], [i[1] for i in arr])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Stochastic Gradient Descent')
plt.show()
```

```python
#mini batch gradient descent
import numpy as np
import matplotlib.pyplot as plt

# Generating synthetic data
x = np.random.rand(100, 1)  # 100 random numbers between 0 and 1
y = 2 * x + 1 + np.random.randn(100, 1)  # y = 2x + 1 + noise

# Initializing parameters
w = np.random.rand(1)
b = np.random.rand(1)
learning_rate = 0.01
batch_size = 10  # Mini-batch size

# Defining the mini-batch gradient descent update rule
def minibatch_update(x_batch, y_batch, w, b, learning_rate):
    n = len(x_batch)
    yhat = w * x_batch + b
    error = y_batch - yhat
    dw = -2 / n * np.sum(x_batch * error)
    db = -2 / n * np.sum(error)
    w = w - learning_rate * dw
    b = b - learning_rate * db
    cost = np.mean(error ** 2)
    return w, b, cost

# Training using Mini-batch Gradient Descent
epochs = 50
arr = []

for epoch in range(epochs):
    # Shuffle the data at the beginning of each epoch
    shuffled_indices = np.random.permutation(len(x))
    x_shuffled = x[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    for i in range(0, len(x), batch_size):
        x_batch = x_shuffled[i:i + batch_size]
        y_batch = y_shuffled[i:i + batch_size]
        w, b, cost = minibatch_update(x_batch, y_batch, w, b, learning_rate)
    
    # Calculate the total loss for the current epoch
    yhat = w * x + b
    loss = np.mean((y - yhat) ** 2)
    print(f'Epoch {epoch} loss is {loss}, parameters are w = {w}, b = {b}')
    arr.append([epoch, loss])

# Plotting the loss over epochs
plt.plot([i[0] for i in arr], [i[1] for i in arr])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Mini-batch Gradient Descent')
plt.show()
```
